```python
import pandas as pd
df =pd.read_csv('../data/ipl_matches_clean.csv')

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())
print("\nMissing values:")
print(df.isnull().sum())
print("\nTeams:")
print(df['team1'].unique())
print("\nTotal matches:", len(df))
```

```python
print("Missing values:")
print(df.isnull().sum())
```

```python
print("All unique teams:")
for team in sorted(df['team1'].unique()):
    print(team)
```

```python
#old name:new name
team_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru'
}

for col in ['team1', 'team2', 'toss_winner', 'winner']:
    df[col] = df[col].replace(team_mapping)

print("After fixing — unique teams:")
for team in sorted(df['team1'].unique()):
    print(team)
print("\nTotal unique teams:", df['team1'].nunique())
```

```python
print("Winner missing/tie matches:")
print(df['winner'].isnull().sum())

print("\nMatch type unique values:")
print(df['match_type'].unique())

print("\nValue counts of winner:")
print(df['winner'].value_counts())
```

```python
# "No Result" wale matches hatao
df = df[df['winner'] != 'No Result']

print("New shape:", df.shape)
print("\nWinner value counts (cleaned):")
print(df['winner'].value_counts())
```

```python
df.to_csv('../data/ipl_matches_final.csv', index=False)
print("Saved! Shape:", df.shape)
```

```python
# Toss winner vs Match winner relationship 
toss_match_same = (df['toss_winner'] == df['winner']).sum()
print(f"Toss winner ne match jeeta: {toss_match_same} times")
print(f"Percentage: {toss_match_same/len(df)*100:.2f}%")
```

```python
print("Top 10 venues by matches played:")
print(df['venue'].value_counts().head(10))
```

```python
# Venue names se city wala extra part hatao
def clean_venue(venue):
    # Comma ke baad ka part agar city ka naam hai to hata do
    if ',' in venue:
        parts = venue.split(',')
        return parts[0].strip()
    return venue.strip()

df['venue'] = df['venue'].apply(clean_venue)

print("After cleaning — Top 10 venues:")
print(df['venue'].value_counts().head(10))
print("\nTotal unique venues:", df['venue'].nunique())
```

```python
df.to_csv('../data/ipl_matches_final.csv', index=False)
print("Final shape:", df.shape)
print("Saved successfully!")
```

```python
# Toss decision (bat/field) vs winning
print("Toss decision distribution:")
print(df['toss_decision'].value_counts())

print("\nWin rate by toss decision:")
toss_decision_win = (df['toss_winner'] == df['winner']).groupby(df['toss_decision']).mean()
print(toss_decision_win)
```

```python
# Target: team1 jeeta ya nahi (1 = haan, 0 = nahi)
df['team1_won'] = (df['winner'] == df['team1']).astype(int)

print(df[['team1', 'team2', 'winner', 'team1_won']].head(10))
print("\nTarget distribution:")
print(df['team1_won'].value_counts())
```

```python
from sklearn.preprocessing import OneHotEncoder

# Features select karo
features = ['team1', 'team2', 'venue', 'toss_winner', 'toss_decision']
X = df[features]
y = df['team1_won']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
print("\nSample features:")
print(X.head())
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop='first')
X_encoded = encoder.fit_transform(X)

print("Original shape:", X.shape)
print("Encoded shape (with drop='first'):", X_encoded.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
print("\nTraining target distribution:")
print(pd.Series(y_train).value_counts())
print("\nTesting target distribution:")
print(pd.Series(y_test).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

print("Model trained successfully!")
print("Number of features model ne dekhe:", X_train.shape[1])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy*100:.2f}%")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nDetailed Report:")
print(classification_report(y_test, y_pred))
```

```python
# Pehle data ko date wise sort karo
df = df.sort_values('date').reset_index(drop=True)

# Function — kisi bhi team ka given match se pehle ka win rate nikalo
def get_team_win_rate_before(team, match_date, df):
    past_matches = df[(df['date'] < match_date) &
                       ((df['team1'] == team) | (df['team2'] == team))]
    if len(past_matches) == 0:
        return 0.5  # naya team, default 50%
    wins = (past_matches['winner'] == team).sum()
    return wins / len(past_matches)

# Test karo ek team pe
sample_rate = get_team_win_rate_before('Mumbai Indians', '2020-01-01', df)
print(f"Mumbai Indians win rate before 2020-01-01: {sample_rate:.2%}")
```

```python
# Har row ke liye team1 aur team2 ka win rate calculate karo
team1_win_rates = []
team2_win_rates = []

for idx, row in df.iterrows():
    t1_rate = get_team_win_rate_before(row['team1'], row['date'], df)
    t2_rate = get_team_win_rate_before(row['team2'], row['date'], df)
    team1_win_rates.append(t1_rate)
    team2_win_rates.append(t2_rate)

df['team1_win_rate'] = team1_win_rates
df['team2_win_rate'] = team2_win_rates

print("Sample data with new features:")
print(df[['team1', 'team2', 'team1_win_rate', 'team2_win_rate', 'winner']].head(10))
```

```python
print("Recent matches (better history available):")
print(df[['date','team1', 'team2', 'team1_win_rate', 'team2_win_rate', 'winner']].tail(10))
```

```python
# Head-to-head function — team1 vs team2 ka specific history
def get_head_to_head(team1, team2, match_date, df):
    past_matches = df[(df['date'] < match_date) &
                       (((df['team1'] == team1) & (df['team2'] == team2)) |
                        ((df['team1'] == team2) & (df['team2'] == team1)))]

    if len(past_matches) == 0:
        return 0.5  # never played before, neutral

    team1_wins = (past_matches['winner'] == team1).sum()
    return team1_wins / len(past_matches)

# Test karo
h2h_rate = get_head_to_head('Mumbai Indians', 'Chennai Super Kings', '2020-01-01', df)
print(f"Mumbai Indians vs Chennai Super Kings (before 2020): {h2h_rate:.2%}")
```

```python
# Har match ke liye head-to-head calculate karo
h2h_rates = []

for idx, row in df.iterrows():
    h2h = get_head_to_head(row['team1'], row['team2'], row['date'], df)
    h2h_rates.append(h2h)

df['head_to_head_rate'] = h2h_rates

print("Sample data with head-to-head:")
print(df[['team1', 'team2', 'team1_win_rate', 'team2_win_rate',
          'head_to_head_rate', 'winner']].tail(10))
```

```python
# Naye features — win rate + head to head + categorical
numerical_features_v4 = ['team1_win_rate', 'team2_win_rate', 'head_to_head_rate']
categorical_features_v4 = ['venue', 'toss_winner', 'toss_decision']

encoder_v4 = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop='first')
cat_encoded_v4 = encoder_v4.fit_transform(df[categorical_features_v4])

X_v4 = np.hstack([df[numerical_features_v4].values, cat_encoded_v4])
y_v4 = df['team1_won']

X_train_v4, X_test_v4, y_train_v4, y_test_v4 = train_test_split(
    X_v4, y_v4, test_size=0.2, random_state=42
)

model_v4 = LogisticRegression(max_iter=1000)
model_v4.fit(X_train_v4, y_train_v4)

y_pred_v4 = model_v4.predict(X_test_v4)
accuracy_v4 = accuracy_score(y_test_v4, y_pred_v4)
print(f"Model v4 Accuracy (with head-to-head): {accuracy_v4*100:.2f}%")
print(f"Previous v3 accuracy was: 53.36%")
```

```python
def get_recent_form(team, match_date, df, n=5):
    past_matches = df[(df['date'] < match_date) &
                       ((df['team1'] == team) | (df['team2'] == team))]

    if len(past_matches) == 0:
        return 0.5

    # Sirf last N matches lo
    recent = past_matches.sort_values('date').tail(n)
    wins = (recent['winner'] == team).sum()
    return wins / len(recent)

# Test karo
recent_form = get_recent_form('Mumbai Indians', '2023-01-01', df)
print(f"Mumbai Indians recent form (before 2023): {recent_form:.2%}")
```

```python
# Har match ke liye recent form calculate karo (last 5 matches)
team1_recent_form = []
team2_recent_form = []

for idx, row in df.iterrows():
    t1_form = get_recent_form(row['team1'], row['date'], df, n=5)
    t2_form = get_recent_form(row['team2'], row['date'], df, n=5)
    team1_recent_form.append(t1_form)
    team2_recent_form.append(t2_form)

df['team1_recent_form'] = team1_recent_form
df['team2_recent_form'] = team2_recent_form

print("Sample with recent form:")
print(df[['team1', 'team2', 'team1_win_rate', 'team1_recent_form',
          'team2_win_rate', 'team2_recent_form']].tail(10))
```

```python
numerical_features_v6 = ['team1_win_rate', 'team2_win_rate', 'head_to_head_rate',
                          'team1_recent_form', 'team2_recent_form']
categorical_features_v6 = ['venue_grouped', 'toss_winner', 'toss_decision']

encoder_v6 = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop='first')
cat_encoded_v6 = encoder_v6.fit_transform(df[categorical_features_v6])

X_v6 = np.hstack([df[numerical_features_v6].values, cat_encoded_v6])
y_v6 = df['team1_won']

X_train_v6, X_test_v6, y_train_v6, y_test_v6 = train_test_split(
    X_v6, y_v6, test_size=0.2, random_state=42
)

model_v6 = LogisticRegression(max_iter=1000)
model_v6.fit(X_train_v6, y_train_v6)

y_pred_v6 = model_v6.predict(X_test_v6)
accuracy_v6 = accuracy_score(y_test_v6, y_pred_v6)
print(f"Model v6 Accuracy (with recent form): {accuracy_v6*100:.2f}%")

# Cross-validation bhi karo — real accuracy janne ke liye
cv_scores_v6 = cross_val_score(model_v6, X_v6, y_v6, cv=5)
print(f"Cross-validation Mean Accuracy: {cv_scores_v6.mean()*100:.2f}%")
```
