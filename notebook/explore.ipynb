```python
import pandas as pd
df =pd.read_csv('../data/ipl_matches_clean.csv')

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())
print("\nMissing values:")
print(df.isnull().sum())
print("\nTeams:")
print(df['team1'].unique())
print("\nTotal matches:", len(df))
```

```python
print("Missing values:")
print(df.isnull().sum())
```

```python
print("All unique teams:")
for team in sorted(df['team1'].unique()):
    print(team)
```

```python
#old name:new name
team_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Deccan Chargers': 'Sunrisers Hyderabad',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru'
}

for col in ['team1', 'team2', 'toss_winner', 'winner']:
    df[col] = df[col].replace(team_mapping)

print("After fixing — unique teams:")
for team in sorted(df['team1'].unique()):
    print(team)
print("\nTotal unique teams:", df['team1'].nunique())
```

```python
print("Winner missing/tie matches:")
print(df['winner'].isnull().sum())

print("\nMatch type unique values:")
print(df['match_type'].unique())

print("\nValue counts of winner:")
print(df['winner'].value_counts())
```

```python
# "No Result" wale matches hatao
df = df[df['winner'] != 'No Result']

print("New shape:", df.shape)
print("\nWinner value counts (cleaned):")
print(df['winner'].value_counts())
```

```python
df.to_csv('../data/ipl_matches_final.csv', index=False)
print("Saved! Shape:", df.shape)
```

```python
# Toss winner vs Match winner relationship 
toss_match_same = (df['toss_winner'] == df['winner']).sum()
print(f"Toss winner ne match jeeta: {toss_match_same} times")
print(f"Percentage: {toss_match_same/len(df)*100:.2f}%")
```

```python
print("Top 10 venues by matches played:")
print(df['venue'].value_counts().head(10))
```

```python
# Venue names se city wala extra part hatao
def clean_venue(venue):
    # Comma ke baad ka part agar city ka naam hai to hata do
    if ',' in venue:
        parts = venue.split(',')
        return parts[0].strip()
    return venue.strip()

df['venue'] = df['venue'].apply(clean_venue)

print("After cleaning — Top 10 venues:")
print(df['venue'].value_counts().head(10))
print("\nTotal unique venues:", df['venue'].nunique())
```

```python
df.to_csv('../data/ipl_matches_final.csv', index=False)
print("Final shape:", df.shape)
print("Saved successfully!")
```

```python
# Toss decision (bat/field) vs winning
print("Toss decision distribution:")
print(df['toss_decision'].value_counts())

print("\nWin rate by toss decision:")
toss_decision_win = (df['toss_winner'] == df['winner']).groupby(df['toss_decision']).mean()
print(toss_decision_win)
```

```python
# Target: team1 jeeta ya nahi (1 = haan, 0 = nahi)
df['team1_won'] = (df['winner'] == df['team1']).astype(int)

print(df[['team1', 'team2', 'winner', 'team1_won']].head(10))
print("\nTarget distribution:")
print(df['team1_won'].value_counts())
```

```python
from sklearn.preprocessing import OneHotEncoder

# Features select karo
features = ['team1', 'team2', 'venue', 'toss_winner', 'toss_decision']
X = df[features]
y = df['team1_won']

print("Features shape:", X.shape)
print("Target shape:", y.shape)
print("\nSample features:")
print(X.head())
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop='first')
X_encoded = encoder.fit_transform(X)

print("Original shape:", X.shape)
print("Encoded shape (with drop='first'):", X_encoded.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
print("\nTraining target distribution:")
print(pd.Series(y_train).value_counts())
print("\nTesting target distribution:")
print(pd.Series(y_test).value_counts())
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

print("Model trained successfully!")
print("Number of features model ne dekhe:", X_train.shape[1])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy*100:.2f}%")

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nDetailed Report:")
print(classification_report(y_test, y_pred))
```

```python
# Team ka overall win rate calculate karo (historical strength)
team_win_rate = df['winner'].value_counts() / (df['team1'].value_counts() + df['team2'].value_counts())
print("Team Win Rates:")
print(team_win_rate.sort_values(ascending=False))
```

```python
print("Matches played by each team:")
matches_played = df['team1'].value_counts() + df['team2'].value_counts()
print(matches_played.sort_values(ascending=False))
```

```python
# Pehle data ko date wise sort karo
df = df.sort_values('date').reset_index(drop=True)

# Function — kisi bhi team ka given match se pehle ka win rate nikalo
def get_team_win_rate_before(team, match_date, df):
    past_matches = df[(df['date'] < match_date) &
                       ((df['team1'] == team) | (df['team2'] == team))]
    if len(past_matches) == 0:
        return 0.5  # naya team, default 50%
    wins = (past_matches['winner'] == team).sum()
    return wins / len(past_matches)

# Test karo ek team pe
sample_rate = get_team_win_rate_before('Mumbai Indians', '2020-01-01', df)
print(f"Mumbai Indians win rate before 2020-01-01: {sample_rate:.2%}")
```

```python
# Har row ke liye team1 aur team2 ka win rate calculate karo
team1_win_rates = []
team2_win_rates = []

for idx, row in df.iterrows():
    t1_rate = get_team_win_rate_before(row['team1'], row['date'], df)
    t2_rate = get_team_win_rate_before(row['team2'], row['date'], df)
    team1_win_rates.append(t1_rate)
    team2_win_rates.append(t2_rate)

df['team1_win_rate'] = team1_win_rates
df['team2_win_rate'] = team2_win_rates

print("Sample data with new features:")
print(df[['team1', 'team2', 'team1_win_rate', 'team2_win_rate', 'winner']].head(10))
```

```python
print("Recent matches (better history available):")
print(df[['date','team1', 'team2', 'team1_win_rate', 'team2_win_rate', 'winner']].tail(10))
```

```python
# Naye features select karo — sirf win rates use karo (simpler model)
features_v2 = ['team1_win_rate', 'team2_win_rate']
X_v2 = df[features_v2]
y_v2 = df['team1_won']

# Train-test split
X_train_v2, X_test_v2, y_train_v2, y_test_v2 = train_test_split(
    X_v2, y_v2, test_size=0.2, random_state=42
)

# Model train karo
model_v2 = LogisticRegression(max_iter=1000)
model_v2.fit(X_train_v2, y_train_v2)

# Accuracy check karo
y_pred_v2 = model_v2.predict(X_test_v2)
accuracy_v2 = accuracy_score(y_test_v2, y_pred_v2)
print(f"New Model Accuracy: {accuracy_v2*100:.2f}%")
```

```python
# Win rate + venue + toss decision combine karo
from sklearn.preprocessing import OneHotEncoder

categorical_features = ['venue', 'toss_winner', 'toss_decision']
numerical_features = ['team1_win_rate', 'team2_win_rate']

encoder_v3 = OneHotEncoder(sparse_output=False, handle_unknown='ignore', drop='first')
cat_encoded = encoder_v3.fit_transform(df[categorical_features])

X_v3 = np.hstack([df[numerical_features].values, cat_encoded])
y_v3 = df['team1_won']

X_train_v3, X_test_v3, y_train_v3, y_test_v3 = train_test_split(
    X_v3, y_v3, test_size=0.2, random_state=42
)

model_v3 = LogisticRegression(max_iter=1000)
model_v3.fit(X_train_v3, y_train_v3)

y_pred_v3 = model_v3.predict(X_test_v3)
accuracy_v3 = accuracy_score(y_test_v3, y_pred_v3)
print(f"Combined Model Accuracy: {accuracy_v3*100:.2f}%")
```

```python
from sklearn.ensemble import RandomForestClassifier

model_rf = RandomForestClassifier(n_estimators=100, random_state=42)
model_rf.fit(X_train_v3, y_train_v3)

y_pred_rf = model_rf.predict(X_test_v3)
accuracy_rf = accuracy_score(y_test_v3, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf*100:.2f}%")
```
